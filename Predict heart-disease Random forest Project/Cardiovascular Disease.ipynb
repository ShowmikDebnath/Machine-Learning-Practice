# Feature Selection (Univarite selection)

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import f_classif
```

```python
FIT_FEATURES = SelectKBest(score_func=f_classif)
```

```python
FIT_FEATURES.fit(x,y)
```

```python
SCORE_COL = pd.DataFrame(FIT_FEATURES.scores_, columns = ['score value'])
```

```python
SCORE_COL
```

```python
NAME_COL = pd.DataFrame(x.columns)
```

```python
topr_features = pd.concat([NAME_COL,SCORE_COL], axis=1)
```

```python
topr_feature = pd.concat([NAME_COL,SCORE_COL],axis=1)
```

```python
topr_feature
```

```python
topr_feature.nlargest(8,'score value')
```

# Feature Importance


```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
model = ExtraTreesClassifier()
```

```python
model.fit(x,y)
```

```python
model.feature_importances_
```

```python
top = pd.Series(model.feature_importances_,index=x.columns)
```

```python
top
```

```python
top.nlargest(10).plot(kind='barh')
```

```python
top.nlargest(10).plot(kind='pie')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=.30,random_state=1)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
Rclf = RandomForestClassifier()
```

```python
Rclf.fit(xtrain,ytrain)
```

```python
Rclf.score(xtest,ytest)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```
